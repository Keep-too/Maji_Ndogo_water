# Assess the quality of water sources:

Write a query to find records where the subject_quality_score is 10 -- only looking for home taps -- and where the source
was visited a second time.

```python
%%sql
SELECT 
	COUNT(*)
FROM md_water_services.water_quality
WHERE subjective_quality_score = 10
AND visit_count = 2;
```

write a query that checks if the results is Clean but the biological column is > 0.01.

```python
%%sql
SELECT 
	source_id,
    description,
    pollutant_ppm,
    biological,
    results
FROM md_water_services.well_pollution
WHERE biological >0.01
LIMIT 20;
```

# Find descriptions ,search for word clean with additional characters after it

```python
%%sql
SELECT 
	*
FROM md_water_services.well_pollution
WHERE description LIKE 'Clean_%';
```

```python
%%sql
SELECT 
	count(*)
FROM md_water_services.well_pollution
WHERE description LIKE 'Clean_%';
```

```python
%%sql 
UPDATE well_pollution
SET description = 
    CASE 
        WHEN description = 'Clean Bacteria: E. coli' THEN 'Bacteria: E. coli'
        WHEN description = 'Clean Bacteria: Giardia Lamblia' THEN 'Bacteria: Giardia Lamblia'
        ELSE description  -- leave unchanged if not matched
    END
WHERE description IN (
    'Clean Bacteria: E. coli',
    'Clean Bacteria: Giardia Lamblia'
);
```
